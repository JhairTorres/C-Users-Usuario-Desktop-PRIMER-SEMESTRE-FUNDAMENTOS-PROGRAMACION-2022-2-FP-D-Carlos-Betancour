```python
#LECTURA DE DATOS
acu_imp=0.0
acu_salario=0.0
cant_emp=int(input('Ingrese cantidad de empleados: '))
for i in range(cant_emp):

  nombre_empleado=input('Ingrese el nombre de el empleado: ')
  num_horas=float(input("Ingrese las horas trabajadas a la semana: "))
  tarifa_hora=float(input("Ingres el salario basico por hora: "))


  if(num_horas<=35):
    salario=num_horas*tarifa_hora
    print("El salario semanal es",salario)
  else:
    salario= 35*tarifa_hora+(num_horas-35)*tarifa_hora*1.5
    print("El salario semanal recargado es",salario)


  if salario<=2000:
    impuestos=0
    print("El total de impuestos es: ",impuestos)
  elif salario<=2200:
    impuestos= (salario-2000)*0.20
    print("El total de impuestos es: ",impuestos)
  elif salario > 2220:
    impuestos=(salario-2220)*0.30 + (220*0.20)
    print("El total de impuestos es: ",impuestos)
    acu_imp= acu_imp+impuestos

  salario_neto=salario-impuestos
  print("El salario neto semanal es de: ",salario_neto)
  acu_salario=acu_salario+salario_neto

  #fin del ciclo
print("Los impuestos a pagar son: ",acu_imp)
print("el valor de los salarios es: ",acu_salario)
  
```
