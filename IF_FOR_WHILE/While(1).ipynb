```python
#Anidado con while
tab=int(input('Ingresa la primera tabla:'))
lim=int(input('Ingresa la ultima tabla:'))
mul=int(input("Ingrese el maximo del multiplicador"))
while tab<lim:
  tab=tab+1
  print('Tabla del:',tab)
  for x in range(mul+1):
    resul=tab*x
    print(tab,'*',x,'=',resul)
print("Fin de la tabla")
```
