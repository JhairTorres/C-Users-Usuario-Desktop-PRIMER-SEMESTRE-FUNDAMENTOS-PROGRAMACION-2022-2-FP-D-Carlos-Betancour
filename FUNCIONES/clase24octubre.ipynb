```python
import os

def menu():
  os.system('clear')
  print('Seleccione una opcion')
  print('\t1 - primera opcion')
  print('\t2 - segunda opcion')
  print('\t3 - tercera opcion')
  print('\t9 - salir')

while True:
  menu()
  opcionmenu = input('Insertar un numero valor>>')

  if opcionmenu=='1':
    print('')
    input('Has pulsado la opcion 1....\npulsa una tecla para continuar')
  elif opcionmenu=='2':
    print('')
    input('Has pulsado la opcion 2....\npulsa una tecla para continuar')
  elif opcionmenu=='3':
    print('')
    input('Has pulsado la opcion 3....\npulsa una tecla para continuar')
  elif opcionmenu=='9':
    break
  else:
    print('')
    input('No has puesto ninguna opcion correcta....\npulsa una tecla para continuar')
  
```
