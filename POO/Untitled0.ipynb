```python
#calculsr para un rectangulo su area su perimetro y su diagonal
class Rectangulo():
  def __init__(self,p_alto,p_ancho):
    self.alto = p_alto
    self.ancho = p_ancho
  def m_perimetro(self):
    perimetro=(self.alto*2)+(self.ancho*2)
    return perimetro
  def m_area(self):
    area=self.alto*self.ancho
    return area

p_alto=int(input("Alto del rectangulo : "))
p_ancho=int(input("Ancho del rectangulo : "))

obj_rect2=Rectangulo(p_alto,p_ancho)

print("El area es : ",+obj_rect2.m_area())
print("El perimetro es : ",+obj_rect2.m_perimetro())
```
